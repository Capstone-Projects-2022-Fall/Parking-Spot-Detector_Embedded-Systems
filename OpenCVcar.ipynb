```python
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('assets/firstpic.png')
# cv2.imshow("test", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
plt.show()
```

```python
img_blur = cv2.GaussianBlur(src=img_gray, ksize=(9, 9), sigmaX=0)
plt.imshow(img_blur,cmap='gray')
plt.show()
```

```python
cv2.imwrite('output/output.jpg',img_blur)
```

```python
start_point = (100, 50)
color = (255, 0, 0)
end_point = (125, 80)
thickness = 1
   
image = cv2.rectangle(img, start_point, end_point, color, thickness)
```

```python
plt.imshow(image)
```
